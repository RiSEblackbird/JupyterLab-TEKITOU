```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

## Case 1
一次元の線上における物体の運動を考える。線上の運動にはあらゆる摩擦は伴わないものとする。
基準からの変位をx[m]とし、時刻t[s]における質量m[kg]の物体に力F[N]が定常的に加えられるときの運動方程式は次式で表される。

$$
m\frac{d^2 x}{dt^2} = F   \tag{1}
$$

式(1)の係数を整理し、速度をv[m/s]と置いて表現し直す。

$$
\begin{cases}
  \frac{dx}{dt} = v \\
  \frac{dv}{dt} = \frac{F}{m} \\
\end{cases} \tag{2}
$$

式（2）に対して、scipy.integrate.odeintを用いてx及びvの時間変化を求めグラフに描写する。

```python
def func_motion_newton(ver, t):
    dxdt = ver[1]
    dvdt = Force/Mass
    
    return [dxdt, dvdt]

def plot2d(t_list, y_list, t_label, y_label):
    plt.xlabel(t_label)
    plt.ylabel(y_label)
    plt.grid()
    plt.plot(t_list, y_list)
    
    plt.show()
    
if (__name__ == '__main__'):
    t_list = np.linspace(0.0, 10.0, 100)
    v0 = 0.0
    x0 = 0.0
    Force = 15.0
    Mass = 2.5
    m_init = [x0, v0]
    m_list = odeint(func_motion_newton, m_init, t_list)
    print(m_list)
    
    plot2d(t_list, m_list[:, 0], "$t[s]$", "$x(t)[m]$")
    plot2d(t_list, m_list[:, 1], "$t[s]$", "$v(t)[m/s]$")
```
